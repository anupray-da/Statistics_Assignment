1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.
- Qualitative Data:- It is descriptive and represents categories or characteristics. It's non-numerical and often focuses on qualities or attributes.
 - Ex. Gender, Language, Religion, Marital Status etc.
- Quantitative data:- It is numerical and represents measurable quantities. It can be counted or measured and is often used for statistical analysis.
 - Ex. Age, Height, Weight, Marks etc.
- Nominal Scale:- It is simplest level of measurement and here data is categorized without any order or ranking.
 - Ex. Gender, Blood type, Colors, Phone Brands etc.
- Ordinal Scale:- Here data is categorized in a specific order.
 - Ex. Income(low,medium,high), Education Level(12th,UG,PG,Masters)
- Interval Scale:- Here data can be categorized, ranked and evenly spaced. There is no true Zero point.
 - Ex. IQ levels, Temperatures
- Ratio Scale:- Here data can be categorized, ranked, evenly spaced and has a  natural Zero point.
 - Ex. Height, weight, age

2. What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.
- Mean:-
 - It is average value of all the data and calculated by summing all values and dividing by the number of values.
 - When to use: When the data is normally distributed (symmetrical) and there are no extreme outliers.
 - Example: The average height of students in a class.
 - Situation: Calculating the average score on a test.
- Median:-
 - It is the middle value when the data is sorted in ascending order and if there's an even number of values, the median is the average of the two middle values.
 - When to use: When the data is skewed (asymmetrical) or contains outliers, as the median is less affected by extreme values.
 - Example: The median income of households in a city.
 - Situation: Real estate angent analyzing house prices to get a better idea of typical home price.
- Mode:-
 - It is the most frequently appears data in the dataset. Also there can be multiple modes or no mode at all.
 - When to use: When dealing with categorical data or when you want to identify the most common value.
 - Example: The most popular color of cars sold in a year.
 - Situation: Determining the most frequent size of shoes sold in a store.

3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?
 - Dispersion:- It refers to the spread or variability of data points in a dataset. It indicates how much the data values deviate from the central tendency.A higher dispersion means the data is more spread out, while a lower dispersion indicates that the data is more clustered around the center.
 - Variance:- It measures the average squared deviation of each data point from the mean. A higher variance indicates greater dispersion.
 - Standard Deviation:- It is the square root of the variance. A higher Standard Deviation indicates greater dispersion.

4.  What is a box plot, and what can it tell you about the distribution of data?
- A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset, showing the median, quartiles, IQR, whiskers, and outliers.
- It tells about data distribution as below:
 - Central Tendency: The median line within the box shows the central value of the data.
 - Dispersion: The length of the box (IQR) represents the spread or variability of the middle 50% of the data. A shorter box indicates less dispersion, while a longer box indicates greater dispersion.
 - Skewness: The position of the median line within the box and the lengths of the whiskers provide information about the skewness of the distribution.
 - Outliers: Outliers are plotted as individual points beyond the whiskers. They can indicate potential errors or unusual observations in the data.

6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?
 - Skewness is a measure of the asymmetry of a probability distribution.
 - It describes the extent to which a distribution's tail is elongated or skewed to one side.
 - A perfectly symmetrical distribution has zero skewness.
 - It is of two types 1)Right Skewness (Positive Skewness) and 2)Left Skewness (Negative Skewness)
 - Right skewness has a longer tail on the right, while left skewness has a longer tail on the left.
 - Skewness affects the choice of the most appropriate measure of central tendency. Also It can influence the interpretation of measures of data spread, such as standard deviation.
 - Skewness can violate the assumption, potentially leading to inaccurate results of certain statistical tests.

7. What is the interquartile range (IQR), and how is it used to detect outliers?
 - The Interquartile Range (IQR) is a measure of statistical dispersion, representing the spread of the middle 50% of a dataset.
 - It is calculated as the difference between the third quartile (Q3) and the first quartile (Q1). In simpler terms, it's the range between the 75th percentile and the 25th percentile of the data.
 - The IQR is used in conjunction with a rule called the "1.5 × IQR rule" to identify potential outliers. This rule defines outliers as data points that fall outside a specific range calculated using the IQR.
 - IQR is a robust and effective method for outlier detection, particularly in skewed distributions.

8. Discuss the conditions under which the binomial distribution is used.
- The binomial distribution is a discrete probability distribution that describes the probability of obtaining a certain number of successes in a series of independent "experiments" where each experiment has only two possible outcomes(like flipping a coin).
- Conditions for Using the Binomial Distribution are:
 - When there is a predetermined, fixed number of trials or experiments conducted.
 - When each trial must be independent of the others, meaning the outcome of one trial does not affect the outcome of any other trial.
 - When each trial must have only two possible outcomes, typically labeled as "success" and "failure."
 - Lastly when the probability of success must remain the same for each trial.

9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).
- Normal Distribution:-
 - The normal distribution, also known as the Gaussian distribution, is a bell-shaped, symmetrical probability distribution that is widely used in statistics.
 - It is characterized by its mean (μ) and standard deviation (σ).
-  properties:-
  - The distribution is shaped like a bell, with the highest point at the mean. It is symmetrical around the mean, meaning the left and right halves are mirror images.
  - The mean, median, and mode of a normal distribution are all equal and located at the center of the distribution.
  - The total area under the normal distribution curve is equal to 1, representing the probability of all possible outcomes.

- Empirical Rule (68-95-99.7 Rule):-
 - This rule provides a quick way to estimate the probability of data falling within certain ranges of a normal distribution.
 - This rule states that for a normal distribution:
 - Approximately 68% of the data falls within one standard deviation of the mean.
 - About 95% of the data falls within two standard deviations of the mean.
 - Almost all (99.7%) of the data falls within three standard deviations of the mean.

10. Provide a real-life example of a Poisson process and calculate the probability for a specific event.
- The Poisson process:-
 - It is a mathematical model used to describe the probability of a certain number of events occurring within a fixed interval of time or space.
 - It's often used to model events that happen randomly and independently over time
- Real life example (Car Accidents on a Highway):-
 - Let's assume that on a particular stretch of highway, there's an average
of 2 accidents per day (λ = 2). We want to calculate the probability of
having exactly 3 accidents on a given day (k = 3).
- The Poisson probability formula: P(X = k) = (e^(-λ) * λ^k) / k!
 - Here X is the random variable representing the number of accidents
 - k is the specific number of accidents we're interested in (k = 3)
 - λ is the average rate of accidents per day (λ = 2)
 - e is the mathematical constant approximately equal to 2.71828
- Calculation:
 - P(X = k) = (e^(-λ) * λ^k) / k!
 - P(X = 3) = (e^(-2) * 2^3) / 3!
 - P(X = 3) ≈ (0.1353 * 8) / 6
 - P(X = 3) ≈ 0.1804
- The probability of having exactly 3 accidents on that stretch of highway on a given day is approximately 0.1804 or about 18.04%.

11. Explain what a random variable is and differentiate between discrete and continuous random variables.
- Random variable:-
 - A random variable is a variable whose value is a numerical outcome of a random phenomenon. In simpler terms, it's a variable that can take on different values based on chance or randomness.
 - Ex. When rolling a die, the random variable could be the number rolled (1, 2, 3, 4, 5, or 6).
- Discrete Random Variable:-
 - It can only take on a finite number of distinct values or a countably infinite number of values.
 - These values are typically whole numbers.
 - Ex. No of cars passing the point on a given time
- Continuous Random Variable:-
 - It can take on any value within a given range or interval.
 - There are infinitely many possible values within this range.
 - Ex. Height of a person
- Difference:-
 - Discrete Random Variable are finite where as Continuous Random Variable are infinite.
 - Discrete Random Variable are typically integers(whole numbers) while Continuous Random Variable are real numbers.
 - Discrete Random Variable are assigned to individual values but Continuous Random Variable are assigned to intervals of values.

12. Provide an example dataset, calculate both covariance and correlation, and interpret the results.
- Covariance:-
 - Covariance measures the directional relationship between two variables.
 - It indicates whether the variables tend to move in the same direction (positive covariance) or in opposite directions (negative covariance).
 - Formula- Cov(x, y) = Σ[(xi - x̄)(yi - ȳ)] / (n - 1)
 - Here xi and yi are individual data points
 - x̄ and ȳ are the means of x and y, respectively
 - n is the number of data points
- Correlation:-
 - Correlation is a standardized measure of the linear relationship between two variables.
 - It ranges from -1 to +1, where -1 indicates a perfect negative correlation, +1 indicates a perfect positive correlation, and 0 indicates no linear correlation.
 - Formula- Corr(x, y) = Cov(x, y) / (σx * σy)
 - Here Cov(x, y) means the covariance of x,y.
 - σx and σy are the standard deviations of x and y, respectively.
